```python
#Data Cleaning Convert Date Columns
df['order_date'].head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['order_date'] = pd.to_datetime(df['order_date'], format='%m/%d/%Y', errors='coerce')
df['ship_date'] = pd.to_datetime(df['ship_date'], format='%m/%d/%Y', errors='coerce')
```

```python
df.info()
```

```python
df['month'] = df['order_date'].dt.month
df['year'] = df['order_date'].dt.year
```

```python
#Business Analysis
```

```python
total_sales = df['sales'].sum()
total_profit = df['profit'].sum()

print("Total Sales:", total_sales)
print("Total Profit:", total_profit)
```

```python
region_sales = df.groupby('region')['sales'].sum().sort_values(ascending=False)
region_sales
```

```python
df['sales'].dtype
```

```python
df['order_date'].dtypehttps://pandas.pydata.org/pandas-docs/stable/user_guide/indexing.html#returning-a-view-versus-a-copy
```

```python
df = df.copy()
```

```python
df['month'] = df['order_date'].dt.month
```

```python
monthly_sales = df.groupby('month')['sales'].sum()
monthly_sales
```

```python
import matplotlib.pyplot as plt

monthly_sales.plot(kind='line', marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(range(1,13))
plt.show()
```

```python
category_profit = df.groupby('category')['profit'].sum().sort_values(ascending=False)
category_profit
```

```python
category_profit.plot(kind='bar')
plt.title("Profit by Category")
plt.ylabel("Total Profit")
plt.show()
```

```python
top_customers = df.groupby('customer_name')['sales'].sum().sort_values(ascending=False).head(10)
top_customers
```

```python
top_customers.plot(kind='bar')
plt.title("Top 10 Customers by Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
discount_profit = df.groupby('discount')['profit'].mean()
discount_profit
```

```python
discount_profit.plot(kind='line', marker='o')
plt.title("Average Profit vs Discount")
plt.xlabel("Discount")
plt.ylabel("Average Profit")
plt.show()
```

```python
df['profit_margin'] = (df['profit'] / df['sales']) * 100
```

```python
df['profit_margin'].describe()
```

```python
df.groupby('category')['profit_margin'].mean()
```

```python
df.groupby('ship_mode')['profit'].mean()
```

```python
yearly_sales = df.groupby('year')['sales'].sum().sort_index()
yearly_sales
```

```python
yearly_sales.plot(kind='line', marker='o')
plt.title("Yearly Sales Growth")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.show()
```

```python
top_products = df.groupby('product_name')['profit'].sum().sort_values(ascending=False).head(5)
top_products
```

```python
top_products.plot(kind='bar')
plt.title("Top 5 Most Profitable Products")
plt.ylabel("Total Profit")
plt.xticks(rotation=45)
plt.show()
```

```python
loss_products = df.groupby('product_name')['profit'].sum().sort_values().head(5)
loss_products
```

```python
df[['sales','profit','discount','shipping_cost']].corr()
```

```python
sns.heatmap(df[['sales','profit','discount','shipping_cost']].corr(), annot=True)
plt.show()
```
